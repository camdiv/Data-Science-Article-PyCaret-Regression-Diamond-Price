```python
import pandas as pd
import numpy as np
```

```python
from pycaret.datasets import get_data
all_datasets = get_data('index')
```

```python
diamond = get_data('diamond')
```

```python
from pycaret.regression import *
```

```python
seen = diamond.sample(frac=0.8,random_state=200) #random state is a seed value
unseen = diamond.drop(seen.index)
```

```python
seen.shape
```

```python
unseen.shape
```

```python
s = setup(data = seen, target = 'Price', session_id = 2)
```

```python
best = compare_models()
```

```python
best_mae = compare_models(sort = 'MAE')
```

```python
evaluate_model(best)
```

```python
pred1 = predict_model(best, data = unseen)
```

```python
tuned_best = tune_model(best)
```

```python
evaluate_model(tuned_best)
```

```python
pred2 = predict_model(tuned_best, data = unseen)
```

```python
top3 = compare_models(n_select = 3)
```

```python
top3
```

```python
stacker = stack_models(top3)
```

```python
evaluate_model(stacker)
```

```python
pred3 = predict_model(stacker, data = unseen)
```
